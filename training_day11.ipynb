# support vector machine
it is a powerful and supervised machine learning algorithm used for classification task as well as regression tasks. 

Aims

to find the optimal hyperplane to separate the data points of different classes and maximize the marginal hyperplane of the closest data points


```python
from sklearn.datasets import load_breast_cancer
```

```python
df = load_breast_cancer()
```

```python
print(df.DESCR)
```

```python
x = df.data
y = df.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='rbf', C=1.0)
```

kernel - it trasforms the data from low dimension to higher dimension

    linear kernel - when the data points are linearly separated
    polynomial kernel - when the data points are not linearly separable
    RBF kernel (radial basis function - this is the most effective when the 
        data is not linearly separable)
    sigmoid kernel - has one hidden layer.
    custom kernel - user defined kernels

```python
svc.fit(X_train_sc, Y_train)
```

```python
pred = svc.predict(X_test_sc)
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
accuracy_score(Y_test, pred)
```

```python
print(classification_report(Y_test, pred))
```
