```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = fetch_california_housing()
```

```python
df.DESCR
```

```python
x = df.data
y = df.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 42)
```

```python
# feature scaling
sc = StandardScaler()
pf = PolynomialFeatures(degree=2)
```

```python
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.fit_transform(x_test)
x_train_pf = pf.fit_transform(x_train)
x_test_pf = pf.fit_transform(x_test)
```

```python
riggs = Ridge(alpha= 1.0)
riggs2 = Ridge(alpha= 1.0)
lasso = Lasso(alpha=0.1)
lasso2 = Lasso(alpha= 0.1)
lr = LinearRegression()
lr2 = LinearRegression()
```

```python
riggs.fit(x_train_sc,y_train)
```

```python
lasso.fit(x_train_sc,y_train)
```

```python
lr.fit(x_train_sc,y_train)
```

```python
riggs2.fit(x_train_pf,y_train)
```

```python
lasso2.fit(x_train_pf, y_train)
```

```python
lr2.fit(x_train_pf, y_train)
```

```python
y_pred_riggs = riggs.predict(x_test_sc)
y_pred_riggs2 = riggs2.predict(x_test_pf)
```

```python
y_pred_lasso = lasso.predict(x_test_sc)
y_pred_lasso2 = lasso2.predict(x_test_pf)
```

```python
y_pred_lr = lr.predict(x_test_sc)
y_pred_lr2 = lr2.predict(x_test_pf)
```

```python
r2_score_riggs = r2_score(y_test,y_pred_riggs)
r2_score_riggs2 = r2_score(y_test,y_pred_riggs2)
```

```python
r2_score_lasso = r2_score(y_test,y_pred_lasso)
r2_score_lasso2 = r2_score(y_test,y_pred_lasso2)
```

```python
r2_score_lr = r2_score(y_test,y_pred_lr)
r2_score_lr2 = r2_score(y_test,y_pred_lr2)
```

```python
mse_riggs = mean_squared_error(y_test,y_pred_riggs)
mse_riggs2 = mean_squared_error(y_test,y_pred_riggs2)
```

```python
mse_lasso = mean_squared_error(y_test,y_pred_lasso)
mse_lasso2 = mean_squared_error(y_test,y_pred_lasso2)
```

```python
mse_lr = mean_squared_error(y_test,y_pred_lr)
mse_lr2 = mean_squared_error(y_test,y_pred_lr2)
```

```python
print(r2_score_lasso, r2_score_riggs, r2_score_lr)
print(r2_score_lasso2, r2_score_riggs2, r2_score_lr2)
```

```python
print(mse_lasso, mse_riggs, mse_lr)
print(mse_lasso2,mse_riggs2,mse_lr2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
x_train
```

```python
processor = ColumnTransformer(
    transformers=[
        ('sc', StandardScaler(), list(range(x_train.shape[1])))
    ]
)
processor2 = ColumnTransformer(
    transformers=[
        ('pf', PolynomialFeatures(), list(range(x_train.shape[1])))
    ]
)
```

```python
pl = Pipeline(
    steps=[
        ('preprocessor',processor),
        ('regression',LinearRegression)
    ]
)
pl2 = Pipeline(
    steps=[
        ('preprocessor',processor2),
        ('regression',LinearRegression)
    ]
)
```
