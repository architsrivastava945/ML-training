```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
plt.plot(x, marker='o', linestyle='-')
'''
line plot is used to visualize the relationship between numerical values.
marker --> 'o' indicates that each point will be marked with a circle.
linestyle --> '-' indicates that the points will be connected with a solid line.
'''
plt.show()
```

### Bar Chart/graph
#### it is used for comparing qualities by categories

```python
x = ['A', 'B', 'C', 'D']
y = [30, 35, 40, 45]

plt.bar(x, y, color='red', width=0.5)
''' 
bar plot is used to visualize the relationship between categorical variables and numerical values.
color --> 'red' sets the color of the bars to red.
width --> 0.4 sets the width of each bar to 0.4 units.
'''
plt.xlabel('Categories') # X-axis label
plt.ylabel('Values') # Y-axis label
plt.title('Bar Chart Example') # Title of the chart
plt.show()
```

```python
plt.barh(x, y, color='blue', height=0.5)
plt.xlabel('Values') # X-axis label
plt.ylabel('Categories') # Y-axis label 
plt.title('Horizontal Bar Chart Example') # Title of the chart
plt.show()
```

### Histogram
### it is used for showing the frequency distribution of a dataset

```python
import numpy as np

data = np.random.randn(1000)  # Generate 1000 random numbers from a normal distribution
print(data)

plt.hist(data, bins=30, color='green', alpha=0.7)
''' 
histogram is used to visualize the distribution of numerical data.
bins --> 30 specifies the number of bins to divide the data into.
color --> 'green' sets the color of the bars to green.
alpha --> 0.7 sets the transparency of the bars to 70%.
'''
plt.xlabel('Value') # X-axis label
plt.ylabel('Frequency') # Y-axis label  
plt.title('Histogram Example') # Title of the chart
plt.grid(alpha=0.75)  # Add grid lines for better readability
plt.show()
```

### Pie chart
### it is used to store the percentage or portion of data

```python
xdata = ['C++', 'Python', 'Java', 'JavaScript']
x = [40, 5, 20, 35]
plt.pie(x, labels=xdata, autopct='%1.1f%%', startangle=140)
''' 
pie chart is used to visualize the proportion of different categories in a dataset.
labels --> xdata specifies the labels for each slice of the pie chart.
autopct --> '%1.1f%%' formats the percentage displayed on each slice to one decimal place.
startangle --> 140 rotates the start of the pie chart to 140 degrees for better visualization.
'''
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is a circle.
plt.title('Pie Chart Example') # Title of the chart
plt.show()
```

### Scatter Plot
### it is used to show the relation between two variables

```python
x = [12, 4, 6, 17, 13, 27, 24, 41]
y = [28, 6, 16, 5, 10, 20, 15, 25]
plt.scatter(x, y, color='purple', marker='x')
'''
scatter plot is used to visualize the relationship between two numerical variables.
color --> 'purple' sets the color of the points to purple.
marker --> 'x' specifies that each point will be marked with an 'x'.
'''
plt.xlabel('x data') # X-axis label
plt.ylabel('y data') # Y-axis label
plt.title('Scatter Plot Example') # Title of the chart
plt.show()
```

### Box plot (whister plot):
### it shows the Statistics summary like median, Quartile and outliers

```python
k = [24, 13, 17, 20, 15, 10, 5, 6, 205]
# 205 is an outlier
plt.boxplot(k, vert=True, patch_artist=True)
'''
box plot is used to visualize the distribution of a dataset and identify outliers.
vert --> True indicates that the box plot will be vertical.
patch_artist --> True allows the box plot to be filled with color.
'''
plt.xlabel('Categories') # X-axis label
plt.ylabel('Values') # Y-axis label
plt.title('Box Plot Example') # Title of the chart
plt.show()
```

### Multiple Line Plot

```python
x = [5, 4, 12, 15, 40]
y1 = [10, 20, 30, 40, 50]
y2 = [4, 50, 70, 20, 18]
y3 = [40, 12, 15, 8, 80]
plt.plot(x, y1, marker='o', color='orange')
plt.plot(x, y2, marker='o', color='red')
plt.plot(x, y3, marker='o', color='blue')
plt.xlabel('X-axis') # X-axis label
plt.ylabel('Y-axis') # Y-axis label
plt.title('multiple Line Plot Example') # Title of the chart
plt.legend(['x, y1', 'x, y2', 'x, y3'])  # Add a legend to differentiate
plt.show()
```

# Seaborn 
## it is a data visualization python library based on matplotlib. it provides high level and attractive statisticals graphs

```python
pip install seaborn
```

```python
import seaborn as sns
```

```python
tip = sns.load_dataset('tips')  # Load the 'tips' dataset from seaborn
# load_dataset() is used to load a dataset from seaborn's built-in datasets.
# tips is a built-in dataset in seaborn that contains information about restaurant tips.
```

```python
tip.head()  # Display the first few rows of the dataset
```

```python
iris = sns.load_dataset('iris')  # Load the 'iris' dataset from seaborn
# iris is a built-in dataset in seaborn that contains information about different species of iris flowers.
```

```python
iris.head()  # Display the first few rows of the dataset
```

```python
sns.lineplot(x = 'tip', y = 'total_bill', data = tip)
# sns.lineplot(x = 'tip', y = 'total_bill', data = tip, hue = 'day')
# hue = 'day' adds a color encoding based on the day of the week, allowing for better visualization of trends across different days.
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = tip)
# barplot compares categories of 'day' against the 'total_bill' values, showing the average total bill for each day.
# bar plot is used to visualize the relationship between categorical variables and numerical values.
```

```python
# countplot : it is used to count the occurrences of each category in a categorical variable.

sns.countplot(x = 'day', data = tip)
```

```python
# boxplot : it is used to visualize the distribution of a dataset and identify outliers.
# sns.boxplot(data = tip) # this shows outliers in the entire dataset
sns.boxplot(x = 'day', y = 'total_bill', data = tip) # this shows outliers day-wise
```

```python
# violinplot : it is used to visualize the distribution and probability density of a dataset and identify outliers.
sns.violinplot(x = 'day', y = 'total_bill', data = tip)
# this has box plot in it, so it shows the distribution of the data along with the median and quartiles.
```

```python
# swarm plot : it is used to visualize the relationship between two numerical variables.
# it is use to diplay all the data points on a categorical axis.
# sns.swarmplot(data = tip)
sns.swarmplot(x = 'day', y = 'total_bill', data = tip)
```

```python
# pairplot : it is used to visualize the pairwise relationships between numerical variables in a dataset.

sns.pairplot(iris, hue='species')
# pairplot shows the relationships between all pairs of numerical variables in the iris dataset, with different colors for each species.
# pairplot is useful for exploring the relationships between multiple variables in a dataset.
# hue = 'species' adds a color encoding based on the species of iris flowers, allowing for better visualization of differences between species.
# it is use to plotting pairwise relationships pf dataset.
```

```python
# heatmap : it is used to visualize the correlation between numerical variables in a dataset.

cor = tip.corr()  # Calculate the correlation matrix of the 'tips' dataset
sns.heatmap(cor, annot=True, cmap='coolwarm')
# annot =True adds the correlation values to the heatmap cells.
# cmap='coolwarm' sets the color map to 'coolwarm', which provides a gradient from cool colors to warm colors.
```

```python
# tip has non-numeric columns, so we need to select only numeric columns for correlation.
# cor = tip.select_dtypes(include=['float64', 'int64']).corr()  # Calculate the correlation matrix using only numeric columns
cor = tip.corr(numeric_only=True)  # Calculate the correlation matrix using only numeric columns
sns.heatmap(cor, annot=True, cmap='coolwarm')
```
