## spam

```python
import pandas as pd
```

```python
df = pd.read_csv("spam.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Category'].value_counts()
```

```python
df['Category'] = df['Category'].map({'ham':0,'spam':1})
```

```python
df.head()
```

```python
df.dtypes
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df['Message']
y = df['Category']
```

```python
x_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
```

```python
clf = Pipeline([
    ('vevtor', CountVectorizer()),
    ('Naive', MultinomialNB())
])
```

```python
clf.fit(x_train, y_train)
```

```python
pred = clf.predict(X_test)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, pred))
```

```python
email = {
    "hey mohan, can we get together to watch football game tomorrow",
    "Upto 20% discount on parking, exclusive after just for you. Don't " 
}
```

```python
clf.predict(email)
```

## Iris

```python
df2 = pd.read_csv("Iris.csv")
```

```python
df2.head()
```

```python
df2['Species'].unique()
```

```python
df2['Species'].value_counts()
```

```python
df2 = df2[df2['Species'] != 'Iris-virginica']
```

```python
df2['Species'].unique()
```

```python
df2.isnull().sum()
```

```python
df2.duplicated().sum()
```

```python
df2['Species'].map({'Iris-setosa':0,'Iris-versicolor':1})
```

```python
x = df2.drop('Species', axis=1)
x = x.drop('Id', axis=1)
```

```python
x.columns
```

```python
y = df2['Species']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
clf2 = Pipeline([
('Naive',GaussianNB(var_smoothing=20.53))
])
```

```python
clf2.fit(X_train, Y_train)
```

```python
pred = clf2.predict(X_test)
```

```python
clf2.score(X_test, Y_test)
```

```python
print(classification_report(Y_test, pred))
```
